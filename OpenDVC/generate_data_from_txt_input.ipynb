```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
result_files = [
    "psnr_encoder_basketballpass_eval_output.txt",
    "mssim_encoder_basketballpass_eval_output.txt",
]
```

```python
for filename in result_files:
    with open(filename, "r") as f:
        
        frames = []
        dB = []
        bpp = []
        mode = None
        if ("psnr" in filename):
            mode = "PSNR"
        elif ("mssim" in filename):
            mode = "MS-SSIM"
        
        for line in f:
            if "Frame" in line:
                data = line.split()
                frames.append(int(data[1]))
                dB.append(float(data[4]))
                bpp.append(float(data[7]))
        
#         points = zip(bpp, dB)
#         sorted_points = sorted(points)
#         new_bpp = [point[0] for point in sorted_points]
#         new_dB = [point[1] for point in sorted_points]
#         plt.figure()
#         plt.title("Reproduced Results for the Encoder of the " + mode + " model")
#         plt.plot(new_bpp, new_dB)
#         plt.xlabel("bpp") 
#         plt.ylabel(mode + " (dB)")
#         plt.grid(True)
#         plt.show()

        bpp, dB = zip(*sorted(zip(bpp, dB)))
        
        plt.figure()
        plt.title("Reproduced Results for the Encoder of the " + mode + " model")
        plt.plot(bpp, dB)
        plt.xlabel("bpp") 
        plt.ylabel(mode + " (dB)")
        plt.grid(True)
        plt.show()
```
